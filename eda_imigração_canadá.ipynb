# Primeiros passos (*First Steps*)

```python
# Importando as bibliotecas que serão usadas ao longo da análise
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tabela1 = pd.read_csv("canadian_immigration_data.csv")
tabela1
```

```python
# Obtendo informações preliminares importantes
tabela1.info()
tabela1.describe()
```

```python
tabela1.set_index("Country", inplace=True)
```

```python
# Criando a lista com os anos apresentados na base de dados
anos = list(map(str, range(1980, 2014)))
```

# Comparação entre países: América do Sul (*Comparison between countries: South America*)


```python
imigrantes_brasil = tabela1.loc["Brazil", anos]
imigrantes_brasil
```

```python
imigrantes_argentina = tabela1.loc["Argentina", anos]
imigrantes_argentina
```

```python
imigrantes_colômbia = tabela1.loc["Colombia", anos]
imigrantes_colômbia
```

```python
dados_brasil = pd.DataFrame(imigrantes_brasil)
dados_brasil = dados_brasil.reset_index()
dados_brasil.columns = ["Ano", "Nº Imigrantes"]
```

```python
dados_argentina = pd.DataFrame(imigrantes_argentina)
dados_argentina = dados_argentina.reset_index()
dados_argentina.columns = ["Ano", "Nº Imigrantes"]
```

```python
dados_colômbia = pd.DataFrame(imigrantes_colômbia)
dados_colômbia = dados_colômbia.reset_index()
dados_colômbia.columns = ["Ano", "Nº Imigrantes"]
```

```python
xticks = list(map(str, range(1980, 2014, 5)))
xticks.append("2013")
xticks
```

```python
plt.figure(figsize=(12,6))
plt.plot(dados_brasil["Ano"], dados_brasil["Nº Imigrantes"])
plt.xticks(xticks)
plt.xlabel("Anos")
plt.ylabel("Número de Imigrantes")
plt.title("Imigração do Brasil para o Canadá")
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(dados_argentina["Ano"], dados_argentina["Nº Imigrantes"])
plt.xticks(xticks)
plt.xlabel("Anos")
plt.ylabel("Número de Imigrantes")
plt.title("Imigração da Argentina para o Canadá")
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(dados_colômbia["Ano"], dados_colômbia["Nº Imigrantes"])
plt.xticks(xticks)
plt.xlabel("Anos")
plt.ylabel("Número de Imigrantes")
plt.title("Imigração da Colômbia para o Canadá")
plt.show()
```

```python
tabela2 = pd.DataFrame([dados_brasil["Nº Imigrantes"], dados_argentina["Nº Imigrantes"], dados_colômbia["Nº Imigrantes"]])
tabela2.index = ["Brasil", "Argentina", "Colômbia"]
tabela2 = tabela2.reset_index()
tabela2.columns = ["País"] + anos
tabela2
```

```python
plt.figure(figsize=(12,6))
plt.plot(dados_brasil["Ano"], dados_brasil["Nº Imigrantes"], label="Brasil")
plt.plot(dados_argentina["Ano"], dados_argentina["Nº Imigrantes"], label="Argentina")
plt.plot(dados_colômbia["Ano"], dados_colômbia["Nº Imigrantes"], label="Colômbia")
plt.xticks(xticks)
plt.xlabel("Anos")
plt.ylabel("Número de Imigrantes")
plt.title("Imigração para o Canadá: Comparação entre nações da América do Sul")
plt.legend()
plt.show()
```
